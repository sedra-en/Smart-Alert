```python
!pip install transformers gradio

from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import gradio as gr

# تحميل النموذج والتوكن
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# إنشاء أنبوب التحليل
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# دالة لتحليل النص
def analyze_sentiment(text):
    result = sentiment_pipeline(text)[0]
    label = result['label']
    score = round(result['score'], 2)

    # خريطة التصنيفات مع الإيموجيات
    stars = {
        "1 star": ("نجمة واحدة", "سلبي جداً 😡"),
        "2 stars": ("نجمتان", "سلبي ☹️"),
        "3 stars": ("3 نجمات", "محايد 😐"),
        "4 stars": ("4 نجمات", "إيجابي 🙂"),
        "5 stars": ("5 نجمات", "إيجابي جداً 😄")
    }

    star_text, feeling = stars.get(label, (label, "غير معروف ❓"))

    return f"التصنيف: {star_text} ({feeling})\nدرجة الثقة: {score}"

# واجهة Gradio
demo = gr.Interface(
    fn=analyze_sentiment,
    inputs="text",
    outputs="text",
    title="تحليل مشاعر التعليقات - عربي",
    description="أدخل تعليقك وسيتم تحليله مع عدد النجوم واسم الشعور ورمز تعبيري!"
)

# تشغيل الواجهة
demo.launch()
```
